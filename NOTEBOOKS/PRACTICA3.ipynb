```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar datos
df = pd.read_csv(r"C:\Users\angel\OneDrive\Escritorio\OCTAVO SEM\MINERIA DE DATOS\REPOSITORIO\AOAG_MinDat035\DATASET\players_20.csv")
```

```python
# Extracción la posición principal (primera de la lista)
df['main_position'] = df['player_positions'].str.split(',').str[0]
```

```python
# 1. Distribución de todas las posiciones principales
position_counts = df['main_position'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(position_counts, labels=position_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Posiciones Principales')
plt.tight_layout()
plt.show()
```

```python
# 2. Distribución de atributos por posición
attributes = ['overall', 'potential', 'age']
for attr in attributes:
    plt.figure(figsize=(10,5))
    for pos in df['main_position'].unique():
        sns.histplot(df[df['main_position'] == pos][attr], label=pos, kde=False, bins=15, alpha=0.5)
    plt.title(f'Histograma de {attr} por posición')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', ncol=1)
    plt.xlabel(attr)
    plt.ylabel('Frecuencia')
    plt.tight_layout()
    plt.show()
```

```python
# 3. Comparacion overall por posición principal
plt.figure(figsize=(15,6))
sns.boxplot(x='main_position', y='overall', data=df)
plt.title('Boxplot de Overall por Posición')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 4. Relación entre potencial y overall
plt.figure(figsize=(10,6))
sns.scatterplot(x='overall', y='potential', hue='main_position', data=df, alpha=0.7)
plt.title('Relación entre Overall y Potential')
plt.xlabel('Overall')
plt.ylabel('Potential')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 5. Promedio de pace por posición
mean_pace = df.groupby('main_position')['pace'].mean().sort_values(ascending=False)
mean_pace.plot(kind='bar', figsize=(12,6), color='steelblue')
plt.title('Promedio de Velocidad (Pace) por Posición Principal')
plt.xlabel('Posición')
plt.ylabel('Pace Promedio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
